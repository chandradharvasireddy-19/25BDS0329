```python
import numpy as np
import pandas as pd
df1 = pd.read_csv("winemag.csv")
print(df1.head(10))
```

```python
df2=pd.DataFrame({'StudentID': [1, 3, 5, 7, 9],'Score': [89, 39, 50, 97, 22]})
df3=pd.DataFrame({'StudentID': [2, 4, 6, 8, 10],'Score': [98, 93, 44, 77, 69]})
ver_cont=pd.concat([df2, df3], ignore_index=True)
print(ver_cont)
```

```python
hor_cont=pd.concat([df2, df3],axis=1)
print(hor_cont)
```

```python
#merging
df4=pd.DataFrame({'StudentID':[2,4,6,8,10,12],'Score1':[98, 93, 44, 77, 69, 56]})
df5=pd.DataFrame({'StudentID':[1, 2, 3, 4, 5, 6, 8, 10],'Score2':[39, 93, 49, 44, 55, 78, 97, 87]})
lj=pd.merge(df4,df5,on='StudentID',how="left")
print(lj)
```

```python
#adding a new column
s=pd.DataFrame({'Q': [100, 250, 300, 150],'U': [50, 60, 45, 80]})
s['TotalPrice']=s['Q']*s['U']
print("Sales Data\n")
print(s)
```

```python
#Descriptive Statistics
print("\nDescriptive Statistics:")
print(s.describe())
```
